```python
import pandas as pd
```

```python
df_produtos = pd.read_csv('produtos.csv')
```

```python
df_produtos
```

```python
#Desempenho por categoria
vendas_categorias = df_produtos.groupby('Categoria')['Quantidade'].sum().reset_index()
```

```python
vendas_categorias
```

```python
#Produto mais vendido
produto_mais_vendido = df_produtos.groupby('Produto')['Quantidade'].sum().idxmax()
```

```python
produto_mais_vendido
```

```python
#Total de unidades vendidas
total_unidades = df_produtos.groupby('Produto')['Quantidade'].max()
```

```python
total_unidades
```

```python
#Correlação UNIDADE X PREÇO
correlacao = df_produtos['Quantidade'].corr(df_produtos['Preço'])
```

```python
correlacao
```
